## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head(20)
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.

application_df = application_df.copy()
Application_df = application_df

Application_df.reset_index(drop=True)
App_df = Application_df.drop(['NAME'], axis=1)
App_df.head()
```

```python
# Determine the number of unique values in each column.

App_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"

App_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

application_types_to_replace = ['T9', 'T13', 'T12', 'T2', 'T25', 'T14', 'T29', 'T15', 'T17']

# Replace in dataframe
for app in application_types_to_replace:
   App_df['APPLICATION_TYPE'] = App_df['APPLICATION_TYPE'].replace(app, "Other")

# Check to make sure replacement was successful
App_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
Application_df['CLASSIFICATION'].nunique()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
value_counts = App_df['CLASSIFICATION'].value_counts()
Filtered_Application = value_counts[value_counts > 1]
value_counts.tail(50)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`

classifications_to_replace = ['C2190', 'C4200', 'C2600','C5200','C1370','C1248','C6100','C1820','C1900','C1236','C3700','C2570','C1580','C1245','C2500','C1570','C1283','C2380','C1732','C1728','C2170','C4120','C8210','C2561','C4500','C2150']

# Replace in dataframe
for cls in classifications_to_replace:
    App_df['CLASSIFICATION'] = App_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
App_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`

Dummies = pd.get_dummies(App_df[['AFFILIATION', 'USE_CASE', 'ORGANIZATION' ,'SPECIAL_CONSIDERATIONS']])

App_df = pd.concat([App_df, Dummies], axis=1)

App_df.drop(['AFFILIATION', 'USE_CASE', 'ORGANIZATION', 'SPECIAL_CONSIDERATIONS'], axis=1, inplace=True)
```
